```python
import yfinance as yf
import pyautogui
import matplotlib
import pyperclip
import time

pyautogui.FAILSAFE = False

ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("6mo")

fechamento = dados.Close
fechamento.plot()

maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)

email = "meuemail@email.com"

pyautogui.PAUSE = 2

pyautogui.hotkey("alt", "tab")
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
pyautogui.click(x = 218, y = 191)
pyperclip.copy(email)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

mensagem = f"""
Prezado gestor, 

Seguem, conforme solicitado, as análises dos últimos seis meses como solicitado da ação ${ticker}

Cotação mínima: R${minima}
Cotação máxima: R${maxima}
Cotação atual: R${atual}

Qualquer dúvida, estou à disposição.

Atte,
Alexandre Rogério
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
pyautogui.click(x = 832, y = 687)

# time.sleep(5)
# pyautogui.position()
```
